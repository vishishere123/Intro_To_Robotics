# Potentiometer

A potentiometer, unlike the name suggests, is actually a variable resistor, which can be used to create variable voltages, using the concept of voltage dividers

### Principle of Working:

The potentiometer combines the concept of variable resistance with that of a voltage divider.

The terminals 1 and 3 in the above figure are connected to the power and GND poles respectively. The terminal 2, which is a wiper arm, is actually connected to the voltage outlet.


The wiper arm moves over the circular track. As the wiper moves away from one of the terminals, the resistance between that terminal and the wiper arm increases, while that between the wiper arm and the other terminal decreases. 


The wiper arm is connected to a voltage output. The voltage received at this output can be estimated from the voltage divider relationship: the voltage drop is directly proportional to the ratio of resistance of the strip between the terminal and wiper arm to the total resistance of the arm.

# Light Sensors


A light sensor is an electronic device which outputs voltage in proportion to the amount of light it is exposed to.

Here, we consider two types of light sensors namely a photoresistor and a phototransistor.


### Photoresistor

As the name suggests, a photoresistor is a device whose resistance varies with the amount of light falling on it. Higher is the amount of light falling on the resistor, lesser is its resistance. Therefore, when used in conjunction with a voltage divider, the voltage signal from the photoresistor can be used to estimate the light falling on it. Since it is essentially a resistor, the photoresistor does not have any polarity; it cannot be connected on both ends interchangeably.

Typically, a photoresistor is less precise for estimating light than a phototransistor.


### Phototransistor

A phototransistor is actually a transistor whose conductance depends on the light falling on it. As with any transistor, it has three parts namely, a collector, an emitter, and a base. The ends of the collector and emitter protrude out of the phototransistor, while the base is fully contained inside the phototransistor.  As the light falling on the phototransistor increases in intensity, the conductance of the phototransistor also increases. Since transistors , including phototransistors, are polar in nature, it is very important to determine the correct ends of the photo transistor and connect them accordingly.

Remember these two points while connecting the phototransistor:

1. The only way to identify the polarity of the phototransistor is to look within the transistor.  The bigger plate is the collector plate, while the smaller plate is the emitter plate.

2. The collector must be connected to the + ve side of the circuit (higher voltage), while the emitter must be connected to negative side (-) , lower voltage.
