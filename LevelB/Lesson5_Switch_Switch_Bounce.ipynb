## Switch Bounce

In a push button switch, the ON or OFF position is attained when the contact between the metal plates is established OR removed. The metal plates do not establish (or disconnect) contact instantaneously. On the contrary, the metal plates will bounce off each other many a time, before reaching their final (ON/OFF) positions. Due to this bouncing, the switch gets turned ON and OFF a lot number of times before finally switching ON or OFF. This phenomenon is known as **Switch Bouncing**.




Switch Bouncing is not a problem when the final instrument does not have the capacity to react to the very fast ON/OFF conditions perpetuated by the switch bouncing.  However, the Pi has the capacity to measure very fast ON/OFF and therefore switch bouncing is a problem when Pi interacts with the push button switches.


It is possible to remove the effect of switch bounce, known as switch-debouncing by both physical and/or software means. While the hardware means to carry out switch debouncing are complicated, the software means are easier.



```python
from IPython.display import Image
Image(filename ="Images/5_d.PNG", height = 800, width =800)
```

Based on the figure above, the simplest way to execute software debouncing is to check, add a delay and then again check, as shown in the following code snippet:

```python
while True:
    if GPIO.input(6) == True:
        time.sleep(0.1)
        if GPIO.input(6) == True:
            count = count + 1
```

However, there may be a couple of problems with this approach:
 - ***if the delay is too less*** In this case, the computer checks for the signal too soon, and therefore overestimates the number of button presses.
 - ***if the delay is too large*** In this case, the computer checks for the signal when the switch has already been released, and therefrore misses to register the button presses.
 
 Therefore, by trial and error, the delay value may be finetuned to yield the correct number of switch operations.

## Pull Up, Pull Down, Active High, Active Low


As mentioned in the introduction to this chapter, the computer of an IoT device or a robot interacts physically with the real world. However, all the inputs are converted to voltages before being hashed  by the CPU. And therefore, the voltage signals coming from the input may be of  3 types:

- **HIGH** This happens when the input pin is connected to a HIGH source, which for the case of Pi is 3.3V.
- **LOW**This happens when the input pin is connected to the GND, resulting in a 0 Voltage being read.
- **FLOATING** This happens when the input pin is not connected to anything, and the input voltage can therefore float to any value between HIGH and LOW, causing the CPU to read the input as HIGH or LOW unpredictably.


We make efforts to not let an input pin to be floating type, which means it should, even in natural state, be connected to a HIGH or LOW source.

- If the input pin is connected to a HIGH source through a large (how large?) resistor, then the input pin is said to be **PULLED UP**.

- If the input pin is connected a GND source through a large resistor, then the input pin is said to be **PULLED DOWN**.

***Note*** *Whether the input pin is PULLED HIGH or LOW, is ascertained from its **natural state**, when it does not receive any active input (when no switch is pressed)*

Pine 20 and Pin 21 are connected to the ground in their natural state, so they are  ***PULL DOWN*** . However, when the switch is *active*, they get connected to the *power source*, and therefore they may also be called as ***ACTIVE HIGH*** connections.


This means that the ***ACITVE HIGH*** and ***PULL DOWN*** are same, though one is defined based on the natural configuration while the other is defined based on the configuration when the switch is active.


Similarly, ***PULL UP*** and ***ACTIVE LOW*** are same.

Note that in the above picture, pull down requires a resistor and connection to ground, while pull up requires a resistor and connection to power source. However, the Pi allows us to set an input as pull up and pull down via software itself.


This is accomplished while declaring the relevant pin as an input pin itself, by using the key word "pull_up_down". See below.

```python
GPIO.setup(20, GPIO.IN, pull_up_down = GPIO.PUD_DOWN)
```

Here, we count the number of times a push button switch is pressed.  The code is given below:

```python
import time, RPi.GPIO as GPIO
GPIO.setmode(GPIO.BCM)
GPIO.setup(13, GPIO.IN, pull_up_down = GPIO.PUD_UP)
count = 0
try:
    while True:
    if GPIO.input(13)== False:
        count = count +1
        print(count)
        time.sleep(1)
    time.sleep(0.05)
except KeyboardInterrupt:
    GPIO.cleanup()
```

### Using Push Button Switch as Toggle Switch

The push button switch, as we know, is only active as long as the button is pressed. Once the button is released, the push button switch goes back to its initial state. In this section, we learn how to keep the circuit in the modified state even after the switch has been unpressed. This is equivalent to using the push button switch as a toggle switch. We accomplish by software.


#### Algorithm

-  Store the status of the output pin (which depends on the switch position) as a binary variable (True or False). Initialize it with the natural state of the output .

- Store the status of the switch as a BOOLEAN variable.

-  Check the status of the switch press. That is, check if the switch has been pressed.

-  If the switch has been pressed , then change the value of the output pin variable. Change also the status of the switch variable.  If it has not been pressed do nothing.

- Forward the new value of output pin variable to the output pin.

- Repeat.

The code is given here which controls the piezo speaker through a push button switch.  The variable SPEAKER controls the status of the  output  while the variables new_switch and old_switch control among themselves, the status of switch input.

```python
from IPython.display import Video
Video("Videos/IMG_2655.MOV")
```

```python
import RPi.GPIO as GPIO
import time
GPIO.setmode(GPIO.BCM)
GPIO.setup(13, GPIO.IN, pull_up_down = GPIO.PUD_UP)
GPIO.setup(26,GPIO.OUT)
SPEAKER = False
GPIO.setup(19,GPIO.OUT)
pwm_19 = GPIO.PWM(19,100)
old_switch = True
try:
    while True:
        new_switch = GPIO.input(13)
        if new_switch ==False and old_switch == True:
            SPEAKER = not SPEAKER            
            if SPEAKER:
                pwm_19.start(50)
            else:
                pwm_19.stop()
            time.sleep(0.2)
            print("Switch Press Detected")
        old_switch = new_switch
        time.sleep(0.2)
except KeyboardInterrupt:
    GPIO.cleanup()
        
        
```
