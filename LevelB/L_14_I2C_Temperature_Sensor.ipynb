# I2C

I2C or $I^2C$ is a serial communication interface (like SPI or UART). It has the following salient features:


1. It is a ***synchronous*** way of transferring data.



2. There are only two lines (pins) for  communication: the CLCK and the Data lines. 



3. Since there is only one data line, all the slaves connected to a master communicate with the same data line. To prevent confusion and collision of data, the slaves will have a unique address each. (Compare with the concept of chip select pins in SPI)



4. Since there is only one data lines, the communication between the slaves and master is ***half duplex***. This means that the transmission of data takes place in both the directions ( from master to slave and from slave to master but not simulataneously).

## How to check if I2C device is connected to the Pi correctly:


To check if the I2C device is interfacing correctly with the Pi involves checking the following :

1. Check if the I2C on the Pi is enabled.

2. Check if the I2C sensor is connected correctly with VCC and GND pins.

3. Check if the I2C sensor is connected correctly with SCL and SDA pins.


All these checks are required to ensure that the I2C is functioning properly.

A single command, when run from the Command Line, will carry out these checks

```python
i2cdetect -y 1
```

# BMP Temperature Sensor

The pin out is shown here:

As the diagram shows, there are 4 pins for I2C, including pins for VCC, GND, SCL and SDA.

The VCC pin requires a 3.3V input, making it ideal for interfacing with the RPi.


The temperature sensor needs a module/drive for easy communicating with the Pi.

Once we plug in the sensor and load the module, we can directly start reading the data from the sensor, without declaring any pins as Input or Output (since we have already declared the relevant pin as SCL or SDA).
