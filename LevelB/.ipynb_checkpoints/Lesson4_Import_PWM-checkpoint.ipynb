## Types of Import in Python

As we know, the import method is used to allow the programmer access to functions and methods of other libraries, beside 
the standard library in Python.

Here, we will discuss about three ways  of importing libraries. And we will evaluate each way of import over two parameters namely:

a.  What all from the library is imported into the program.

b.  How are the imported functions called inside the program.


Let us first inspect the simplest way to use the import function, which is:

```python
import time
```

```python
import os
```

In the above type of import:

- Everything inside the module is imported.

- The functions inside the library/module are called as:

```python
time.sleep(3)
```

This means, that function [sleep] is invoked by calling the module 'time' within which it is contained, followed by the '.' symbol.  

The second way to use the import function is by using an alias. This is shown as follows:

```python
import time as tm
```

Here, the module 'time' has been renamed for the purpose of this program, as 'tm', which suggests that in this type of import:


- Everything inside the module is imported.

-  The functions inside the library/module are called as:


```python
tm.sleep(5)
```

Note that this way of importing allows for 'time' module to be referenced as 'tm'.

The third and the last way of importing modules is given as:

```python
from time import sleep
```

In the context of our discussion, the above method of using the 'import' function  means:

-  Only the 'sleep' function inside the 'time' module is imported.

-  The function 'sleep' inside the library/module is called directly as:


```python
sleep(10)
```

This is to say that we do not need to refer to time module to call sleep function.

This (3rd) way to use the 'import' function, can also be used in a broader way as:

```python
from time import *
```

Again, when evaluated against our paradigm for evaluating ways to use 'import', we understand:

a.  Everything in the 'time' module is imported.

b. We can use the methods of 'time' module directly, as in:

```python
sleep(10)
```

This means that "best of both worlds" are combined. We are able to invoke the functions directly, as well as all the functions become available. However, on the flip side, we see that it is difficult to distinguish between such functions (which are called from other modules using import) and the default/native functions of Python.

## Pulse Width Modulation:  PWM

The GPIO (general purpose input and output) pins of Pi can only output **HIGH** and **LOW** voltages namely **3.3V** and **0** Volt respectively. This type of signal is called a **digital** signal. Such signals are useful for , among other things, transmission of data from the internet to your device.

However, digital signals are not useful when there is a requirement for signals which vary across a continuous spectrum, for example the signals required to drive a speaker. For such cases, we require analog signals. 

Since Pi is not able to output analog signals, we need a technique/ paradigm to convert the digital signals to analog signals.  **Pulse Width Modulation (PWM)** is such a paradigm.


### PWM: Concept

 A digital signal by definition has only two values: ***HIGH*** and ***LOW***.  However, if the signal value between the HIGH and LOW is changed at a very high frequency, then the resulting signal may appear like an analog signal.
 
 There are two concepts related to PWM namely:
 
 - Frequency
 - Duty Cycle

Before we understand ***frequency***, we understand cycle. In any (periodic) signal, a cycle is when the voltage comes back to the same value when it started. For example, in a digital signal, the time from beginning of one high to the beginning of another high constitutes one cycle.

**Frequency** then denotes the number of cycles in 1 second.

In the above figure, one cycle is shown, and it has a period of one second, which means the frequency is 1 Hertz.

Here, in the above figure, we see that there are 10 cycles in 1 second, and therefore the frequency is 10 Hz.

**Duty cycle** on the other hand, refers to the percentage of time in a cycle, the signal is HIGH.

```python
from IPython.display import Image
Image(filename ="Images/4_c.png", width = 800, height =800)
```

The above figure shows three (3) signals with 50 %, 10% and 80% duty cycles respectively.

***How to choose duty cycle and frequency for any component?*** The frequency depends on the component and the duty cycle depends on the amount of brightness (LED) or loudness (speaker) you want.

**Speakers and PWM**: Speakers should be run by only analog signals, not digital signals (not even digital signals with PWM). However, there are special types of speakers called piezoelectric speakers which are amenable to PWM digital signals. These speakers are generally used in computers during bootup processes and error flashing. By changing the frequency and duty cycles of the signal being fed into a piezo speaker, different kinds of sounds can be extracted from the piezo speaker.

Among piezo speakers, there is a special type of piezo speakers called buzzer. A buzzer is a piezo speaker where the frequency and the duty cycle are inbuilt inside the hardware itself, and cannot be changed. Therefore, a buzzer outputs only 1 kind of sound, and inputs only 1 kind of signal (digital ON/OFF).

**Hardware vs Software PWM** : The Pulse Width Modulation (PWM) can be incorporated through both hardware and software. In hardware based PWM, there is a time or microprocessor which is responsible for generating the PWM signals. The program (software) only sends the frequency and the duty cycle, the rest is taken care of by the software. However, on the flip side, using hardware based PWM requires specialized hardware, which is missing in less expensive devices. For example, in RPi, only 1 pin namely GPIO18 can be configured as a hardware PWM pin.


On the other hand, in software PWM, the software is responsible for turning the pin on and off repeatedly (at a given frequency with a given duty cycle) to generate the PWM signal. There is clear advantage that several GPIO pins can be configured for PWM signals. On the flip side, software PWM requires more CPU resources than hardware PWM. Having said that, the Pi has enough CPU resources to comfortably use several GPIO pins for software PWM.

## Software PWM

The RPi.GPIO module, which we have been using to interface with RPi, has commands to execute PWM as well.

```python
import RPi.GPIO as GPIO

channel = 19  #Declare the pin number for which we need to carry out PWM analysis.

frequency = 10  #Declare the frequency for the PWM

new_frequency  = 20  # Declare one more frequency for the PWM

duty_cycle = 50  # Declare the duty cycle to be 50 percent

pwm = GPIO.PWM(channel, frequency)  # Use the PWM command in the GPIO module to intiate PWM in a channel with some given frequency.

pwm.start(duty_cycle)  #Starts the PWM modulation with a given duty cycle.

pwm.ChangeFrequency(new_frequency)  # Changes the frequency of the pwm signal.

pwm.ChangeDutyCycle(duty_cycle -20)  # Changes the duty cycle of the pwm signal.

pwm.stop()
```
