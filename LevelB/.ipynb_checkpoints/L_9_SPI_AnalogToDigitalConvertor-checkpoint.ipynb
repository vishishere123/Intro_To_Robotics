## Voltage Divider



A voltage divider is a useful concept in electrical/electronics engineering, which allows us to obtain a source of desired voltage from another source of different voltage using resistors placed in series.

For example, see the figure below:


```python
from IPython.display import Image
Image(filename = "Images/VoltageDivider.PNG", height = 200, width = 200)
```

Some salient (mostly all) features of the voltage divider:

- There can be any number of resistors, not necessarily 2.

- The resistors may not be equal to each other, may all be different.

- **The Voltage Drop across any resistor $R_1$  equals $\Delta V = V \times R_1/(R_1+R_2+R_3+...)$**
, where V is the source voltage and $R_1, R_2, R_3...$ are the resistances.

- The voltage available at the output of the resistor is then:

    $V - \Delta V$

# Analog Signals


We know that the Raspberry Pi is capable of handling only digital signals, i.e. signals which have only two values HIGH and LOW. 

On the other hand, the signals which vary continuously from LOW and HIGH are known as analog signals.

Most sensors, if not all, convey their information through analog signals. For example, a light sensor, whose function is to estimate the intensity of light, sends the information about light intensity through voltages continously varying in proportion to the intensity of light. Similarly, a temperature sensor will also output continuously varying analog signals in proportion to the temperature. The same also holds true for humidity sensors, ultrasonic sensors, infra red sensors and others. Each sensor will output a voltage signal in proportion to the quantity of the variable it is sensing.

This leads to the conclusion that for a Raspberry Pi to be able to understand and read the sensors, it must have the capacity to read analog signals. To allow the Pi to accomplish this task, we must use a type of integrated cicruit (IC) known as A/D convertor (analog to digital convertor).

## Integrated Circuit

## Naming convention of pins

The pins on the IC are named according to the position of the notch. **The pin which is closest to the notch on the left hand side is named as 1**, while the pins are then numbered in a somewhat counter clockwise order. When the pins on the left hand side are exhausted, the numbering continues on the right hand side, from the bottom.

Some pins also have a dot (apart from the notch) to denote Pin 1.

## IC Data sheet

By looking at the IC itself, we can obtain information about the IC name and the pin nomenclature itself. However, to be able to use the IC, we need more information, for example what does the IC do, what is the function of each pin, how much power does the IC require among others. This set of information can only be required by looking at the IC ***data-sheet***.

To obtain the data-shet, just search for the IC name + data sheet on the internet and it will be available. For example, for the IC which we are going to use in this project lesson, the name of the IC is MCP3008. A google search for the term "MCP3008+data sheet" yields the following PDF file:


```python
from IPython.display import IFrame
IFrame("MCP3008.pdf", height = 1200, width =400)
```

From the first 2 pages of the Data-Sheet above, we can see that:

- The IC is a A/D (analog to digital) convertor.

- The IC has a SPI communication interface.

- The IC has a resolution of 10 bits.

- It has 8 input channels.

- It needs supply voltage of 2.7 -5 Volts.


- The pin diagram is provided.


To explain in detail:

- The resolution is 10 bits. This corresponds to 2^10 = 1024. Since the supply voltage will be 3.3V from the Pi, so the resolution is 3.3/1024 = 0.003222 Volts, which is very fine.

- The supply voltage can be anywhere between 2.7 and 5 Volts, which covers the 3.3 Voltages supplied by Pi, and so this IC is perfectly compatible with Pi.


- It has 8 input channels with the pinout provided as follows:



```python
from IPython.display import Image
Image(filename = 'Images/MCP3008_PINOUT.PNG', height = 400, width = 400)
```

- The pins on the left correspond to analog input channels, numbered from CH0 to CH7. 

- The channels on the right  include 2 channel ground AGND and DGND, referring respectively to Analog Ground and Digital ground respectively.

- The two voltage sources channel include V_DD and V_REF.

- The four remaining channels included CLK, DOUT,  DIN and CS/ SHDN.  These are expanded as Clock, Digital Out, Digital In, and ShutDown. Since the IC communicates by SPI interface, which needs 4 pins, these 4 pins will be required for establishing the communication between IC and Pi.

### Digital Communication

There are 3 protocols of digital communication, namely:

-  I2C

- UART

- SPI


This IC, as mentioned in the data-sheet, uses 4 pins for communicating.

- CLK  : Clock for synchronizing the communication between the two devices.

- CE  : Chip enables multiple devices to share MISO, MOSI

-  MISO:  Master In, Slave Out

- MOSI:  Master Out, Slave In 

Here Master is the IC, and RPi is the slave.  MISO means the pin receives the output from the Pi.  MOSI means the pin releases the output from the IC.


As far as the Pi is considered, it is possible to enable SPI communication in the RPi in 2 ways:


- Hardware SPI:   Here, some of the GPIO pins of the RPi are used for SPI. The salient features of this SPI are as follows:

- Hardware SPI utilizes more CPU.

- Hardware SPI requires certain option in the Raspian OS to be switched ON.

- Hardware SPI can be used only in some pre-configured (not random) pins of the RPi.

- Harware SPI is generally faster,  but for small amount of data transfer, the increase in speed afforded by the hardware SPI is not much.


On the other hand, for this project we are going to use Software SPI:

-  Here, the software is used for establishing SPI.


- To use this option, there is no need to change settings in Raspbian, however we need to download (and install/use) a library for the IC.

- We can use any pin for this IC, as long as it is mentioned by the driver.


For this IC, we need to be using the driver for MCP3008 available from Adafruit, named as Adafruit_MCP3008.



### Adafruit_MCP3008

This Python library, helps the RPi to interface with the MCP3008 IC using the SPI communication protocol.

The Python library is installed from the official github library: 

https://github.com/adafruit/Adafruit_Python_MCP3008


This Adafruit MCP3008 library has some pre-configured pins for SPI communication. We could change these pins in the program, but we refrain from doing that to avoid overcomplicating the process.

These, self-configured pins are:

CLK - GPIO18

MISO - GPIO23

MOSI - GPIO24

CS - GPIO25

The pseudo code for using this library is given as





```python
import Adafruit_MCP3008
#Import the library
```

```python
#Declare a MCP3008 object with the key parameters of clk, miso,mosi, cs

mcp1 = Adafruit_MCP3008.MCP3008(clk = 18, miso = 23, mosi = 24, cs= 25)

#Use this object to read the values from channel 0:


value1 = mcp1.read(0)

```
