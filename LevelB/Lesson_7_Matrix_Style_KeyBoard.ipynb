# Matrix Style KeyBoard

The pushbutton switch which we have seen earlier, can control only 1 input while it needs one pin of the computer (Pi). If we have a keyboard with 50+ keys, then using a push button switch for each key would require the computer to have 50+ pins, which are not available, especially in inexpensive devices like Raspberry Pi.

To get around this problem, we use another method of switching, called matrix style switching. The figure is shown below:

```python
from IPython.display import Image
Image(filename = "Images/MatrixStyleKeyboard.PNG", height = 300, width =300)
```

The salient features of the matrix style key board are shown here:

- All the switches in one row are inter-connected, while each row connects to one pin of the Pi.

- Similarly, all the switches in one column are inter-connected, while each colum connects to one pin of the Pi.

- The number of pins required is therefore = No. of Rows + No. of Columns

- The number of switches = No. of Rows * No. of Columns

- Therefore, by using (Rows+Columns) pins, we can use (Rows x Columns) switches.  This affords us to use more number of switches without using a large number of pins

Working of the matrix style keyboard:

- Each switch lies at the connection of 1 row and 1 column.

- When the switch is pressed connection is established between the row and column of that switch.



### Software for the matrix style key pad

Therefore, an algorithm to detect which switch has been pressed shall look like the following:


1. Make Row A HIGH (=3.3V)

2. Check if any of the columns are HIGH. If any of the columns are active, then the switch corresponding to that column and that row become active. If none of the columns are HIGH, move to next Step

3. Make Row A LOW.

4. Make Row B HIGH, and check all the columns are high just like for Row A.

### Wiring of the keypad

As discussed earlier, the number of wires in a matrix style keypad equals the number of Rows +number of columns. Since that equals 7 for our case, the keypad will and does have 7 wires. Each of these wires is connected to each row and column.
